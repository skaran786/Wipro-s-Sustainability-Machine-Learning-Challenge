```python
import numpy as np
import pandas as pd
from thundersvm import SVC
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
```

```python
def scaling(df):
    '''This function will scale the data using Standard Scaler'''
    scale = StandardScaler()
    col = df.columns
    df = pd.DataFrame(scale.fit_transform(df))
    df.columns = col
    return df
```

```python
df = pd.read_csv('/content/drive/MyDrive/WiproCompetetion/train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X_train = df.drop(['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI','Year','Month','Day','Minute','Cloud Type','Pressure','Wind Direction','Fill Flag'], axis=1)
y1 = df['Clearsky DHI']
y2 = df['Clearsky DNI']
y3 = df['Clearsky GHI']
```

```python
X_train = scaling(X_train)
```

```python
from thundersvm.thundersvm import SvmModel
model1 = model2 = model3 = RandomForestRegressor()
```

```python
print("Training model")
print("Model 1 started")
model1.fit(X_train,y1)
print("Model 1 ended")
print("Model 2 started")
model2.fit(X_train,y2)
print("Model 2 ended")
print("Model 3 started")
model3.fit(X_train,y3)
print("Model 3 ended")
print("Training done..")
```

```python
import pickle

filename1 = 'finalized_model1.sav'
filename2 = 'finalized_model2.sav'
filename3 = 'finalized_model3.sav'
pickle.dump(model1, open(filename1, 'wb'))
pickle.dump(model1, open(filename2, 'wb'))
pickle.dump(model1, open(filename3, 'wb'))
```

```python
test = pd.read_csv('/content/drive/MyDrive/WiproCompetetion//test.csv')
test.head()
```

```python
X_test = test.drop(['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI','Year','Month','Day','Minute','Cloud Type','Pressure','Wind Direction','Fill Flag'], axis=1)

X_test = scaling(X_test)
```

```python
import pickle
loaded_model = pickle.load(open('finalized_model1.sav', 'rb'))
result1 = loaded_model.score(X_test,y1)
```

```python
pred1 = model1.predict(X_test).astype('int32')
pred2 = model2.predict(X_test).astype('int32')
pred3 = model3.predict(X_test).astype('int32')
```
