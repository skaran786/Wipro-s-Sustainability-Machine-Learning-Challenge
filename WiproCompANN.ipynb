```python
import pandas as pd
import os
import numpy as np
from glob import glob
from matplotlib import pyplot as plt
import seaborn as sns
from random import randint
from glob import glob
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import mean_squared_error
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv('/content/drive/MyDrive/WiproCompetetion/train.csv')
df.head()
```

```python
y1 = df['Clearsky DHI']
y2=df['Clearsky DNI']
y3=df['Clearsky GHI']
X_train = df.drop(['Clearsky DHI', 'Clearsky DNI', 'Clearsky GHI','Year','Month','Day','Minute','Cloud Type','Wind Direction','Temperature','Fill Flag'], axis=1)
```

```python
model1 = keras.Sequential()
model1.add(layers.Dense(300, activation="relu"))
model1.add(layers.Dense(200, activation="relu"))
model1.add(layers.Dense(100, activation="relu"))
model1.add(layers.Dense(50, activation="relu"))
model1.add(layers.Dense(25, activation="relu"))
model1.add(layers.Dense(1, activation='relu'))
model2 = keras.Sequential()
model2.add(layers.Dense(300, activation="relu"))
model2.add(layers.Dense(200, activation="relu"))
model2.add(layers.Dense(100, activation="relu"))
model2.add(layers.Dense(50, activation="relu"))
model2.add(layers.Dense(25, activation="relu"))
model2.add(layers.Dense(1, activation='relu'))
model3 = keras.Sequential()
model3.add(layers.Dense(300, activation="relu"))
model3.add(layers.Dense(200, activation="relu"))
model3.add(layers.Dense(100, activation="relu"))
model3.add(layers.Dense(50, activation="relu"))
model3.add(layers.Dense(25, activation="relu"))
model3.add(layers.Dense(1, activation='relu'))
```

```python
model1
model2
model3
```

```python
from sklearn.model_selection import train_test_split                    
from sklearn.neural_network import MLPRegressor   
model1 = MLPRegressor(solver='lbfgs',alpha=0.001,hidden_layer_sizes=(150,))
model2 = MLPRegressor(solver='lbfgs',alpha=0.001,hidden_layer_sizes=(150,))
model3 = MLPRegressor(solver='lbfgs',alpha=0.001,hidden_layer_sizes=(150,))
```

```python
model1.fit(X_train,y1)
```

```python
model2.fit(X_train,y2)
```

```python
model3.fit(X_train,y3)
```

```python
model1.score(X_train,y1)
pre=model1.predict(X_train)
```

```python
#Compile and train
model1.compile(optimizer=Adam(learning_rate=0.0001), 
              loss='mse',
              metrics=['mae',keras.metrics.RootMeanSquaredError()])

history = model1.fit(X_train,y1, epochs=100, batch_size=64)
```

```python
#Compile and train
model2.compile(optimizer=Adam(learning_rate=0.0001), 
              loss='mse',
              metrics=['mae',keras.metrics.RootMeanSquaredError()])

history2 = model2.fit(X_train,y2, epochs=100, batch_size=64)
```

```python
model1.evaluate(X_train,y1, verbose=1)
model2.evaluate(X_train,y2, verbose=1)
model3.evaluate(X_train,y3, verbose=1)
```
